IMPORTING NECESSARY LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

FUNCTION TO READ THE CSV FILE

```python
def read_file(file_name):
    file = pd.read_csv(file_name)
    X = file.iloc[:,0]
    Y = file.iloc[:,1]
    plt.scatter(X,Y)
    plt.show()
    return X, Y
```

GRADIENT DESCENT WITHOUT REGULARIZATION PARAMETER

```python
#Gradient descent
def grad_without_regularization(W, b, X, Y, l_r):
    gradient_W = 0
    gradient_b = 0
    y = np.sum(W * X + b)
    m  = len(X)

    gradient_W = (1/m) * np.sum(((W * X + b) - Y) * X)
    gradient_b = (1/m) * np.sum((W * X + b) - Y)
    W = W - l_r * gradient_W
    b = b - l_r * gradient_b
    plt.plot(X, W * X + b, color="red")  # Regression line

    return W, b
```

```python
#Gradient descent
def grad_with_regularization(W, b, X, Y, l_r, ld):
    gradient_W = 0
    gradient_b = 0
    y = np.sum(W * X + b)
    m  = len(X)

    gradient_W = (1/m) * (np.sum(((W * X + b) - Y) * X) + ld * W)
    gradient_b = (1/m) * (np.sum((W * X + b) - Y) + ld * b)
    W = W - l_r * gradient_W
    b = b - l_r * gradient_b
    plt.plot(X, W * X + b, color="red")  # Regression line

    return W, b
```

MAIN FUNCTION

```python
if __name__ == "__main__":
    file_name = 'Salary_dataset.csv' 
    X, Y = read_file(file_name)
    
    regularization_yes1_no0 = 1
    #Initializing W and b to random values
    W = 20
    b = -2

    #Initializing epochs
    e = 30

    #Initializing learning rate
    l_r = 0.01

    #Initializing lambda
    ld = 100

    plt.scatter(X, Y, color="blue")  # Scatter plot of data points
    

    #Computing cost
    if regularization_yes1_no0 == 1:
        print('yes')
        for i in range(e):
            #error = cost(W, b, X, Y)
            W, b = grad_with_regularization(W, b, X, Y, l_r, ld)
            print(W, b)

    else:
        for i in range(e):
            #error = cost(W, b, X, Y)
            W, b = grad_without_regularization(W, b, X, Y, l_r)
            print(W, b)
    
```
